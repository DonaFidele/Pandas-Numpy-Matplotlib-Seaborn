```python
import numpy as np
```

```python
X=np.arange(10)
Y=2*X*X*X
plt.plot(X,Y,'g--')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.show()
```

```python
X=np.arange(10)
Y=np.random.randint(0,100,len(X))
plt.scatter(X,Y,marker='x' ,c='y')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.show()
```

```python
X=['A','B','C','D']
Y=[10,20,30,40]
plt.pie(Y,labels=X)
plt.title('line plot')
plt.show()
```

```python
X=['A','B','C','D']
Y=[10,20,30,40]
plt.bar(x=X,height=Y,width=0.35)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.show()
```

```python
X=['A','B','C','D']
Y1=[10,20,30,40]
Y2=[15,25,35,45]
plt.bar(x=X,height=Y1,bottom=None,width=0.35)
plt.bar(x=X,height=Y2,bottom=Y1,width=0.35)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.show()
```

```python
plt.subplot(2,2,1)
X=['A','B','C','D']
Y=[10,20,30,40]
plt.bar(x=X,height=Y,width=0.35,color='b')

plt.subplot(2,2,2)
X=['E','F','G','H']
Y=[15,25,35,45][::-1]
plt.bar(x=X,height=Y,width=0.35,color='g')

plt.subplot(2,2,3)
X=['X','Y','Z']
Y=[9,15,1]
plt.bar(x=X,height=Y,width=0.35,color='y')

plt.subplot(2,2,4)
X=['X','Y','Z']
Y=[9,15,1][::-1]
plt.bar(x=X,height=Y,width=0.35,color='r')

```
